```python
df1 = df1.rename(index=str, columns={"Restaurant": "Name"})
df_merge = df.merge(df1, on='Name')
df_merge.head()
```

```python
df_merge.drop('Collections',axis=1,inplace=True)
df_merge.head()
```

```python
df_merge['Rating'].value_counts()
```

```python
def convert_review_rating(Rating):
    if Rating=="1.5":
        Rating =1
    elif Rating == "1":
        Rating =1
    elif Rating =="2.5":
        Rating = 2
    elif Rating =="2":
        Rating = 2
    elif Rating =="3.5":
        Rating = 3
    elif Rating =="3":
        Rating = 3
    elif Rating == "4.5":
        Rating = 4
    elif Rating =="4":
        Rating = 4
    elif Rating =="5":
        Rating = 5
    elif Rating =="Like":
        Rating = 5
    return Rating
```

```python
df_merge["Rating"] = df_merge["Rating"].apply(convert_review_rating)
```

```python
df_merge['Rating'].value_counts()
```

```python
plt.figure(figsize=(400,30))

sns.countplot(x="Name",hue="Rating",data=df_merge)

```
